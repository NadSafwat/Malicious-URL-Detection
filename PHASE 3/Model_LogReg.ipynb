```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, classification_report
```

```python
df_train = pd.read_csv(r'C:\Users\nadin\OneDrive\Documents\Machine Learning Project\train_dump.csv')
```

```python
columns_to_convert = df_train.columns[8:-1]
df_train[columns_to_convert] = df_train[columns_to_convert].astype('int8')
```

```python
df_test = pd.read_csv(r'C:\Users\nadin\OneDrive\Documents\Machine Learning Project\test_dump.csv')
```

```python
df_test[columns_to_convert] = df_test[columns_to_convert].astype('int8')
```

```python
df_train = df_train.iloc[:, 1:]
df_test = df_test.iloc[:, 1:]
```

```python
x_train = df_train[df_train.columns[:-1]].to_numpy()
y_train = df_train['label'].to_numpy()

x_test = df_test[df_test.columns[:-1]].to_numpy()
y_test = df_test['label'].to_numpy()
```

```python
kf = KFold(n_splits=5)

f1s = []
accuracies = []
precisions = []
recalls = []
roc_aucs = []
```

```python
KF = KFold(n_splits=5)
def Model_Train(Model, X, Y):
    Preds = []
    fold = 1
    for i,j in KF.split(X):
        print(f"##### FOLD: {fold} #####")
        Train_X = X.iloc[i]
        Test_X = X.iloc[j]
        Train_Y = Y.iloc[i]
        Test_Y = Y.iloc[j]
        
        Model.fit(Train_X, Train_Y)
        Pred_Y = Model.predict(Test_X)
        Preds.append(Pred_Y)
        # Evaluate the model
        precision = precision_score(y_true=Test_Y, y_pred=Pred_Y, zero_division=0)
        recall = recall_score(y_true=Test_Y, y_pred=Pred_Y, zero_division=0)
        accuracy = accuracy_score(y_true=Test_Y, y_pred=Pred_Y)
        f1 = f1_score(y_true=Test_Y, y_pred=Pred_Y, zero_division=0)

        # Store the result
        f1s.append(f1)
        accuracies.append(accuracy)
        precisions.append(precision)
        recalls.append(recall)

        # Print the scores for each fold
        print(f"Precision = {precision}")
        print(f"Recall = {recall}")
        print(f"Accuracy = {accuracy}")
        print(f"F1 score = {f1}\n")
        



        fold += 1
```

```python
model = LogisticRegression()

fold = 1
for train, valid in kf.split(x_train, y_train):
    print(f"##### FOLD: {fold} #####")

    # Fit the model
    model.fit(x_train[train], y_train[train])

    # Predict on the test set
    predictions = model.predict(x_train[valid])

    # Evaluate the model
    precision = precision_score(y_true=y_train[valid], y_pred=predictions, zero_division=0)
    recall = recall_score(y_true=y_train[valid], y_pred=predictions, zero_division=0)
    accuracy = accuracy_score(y_true=y_train[valid], y_pred=predictions)
    f1 = f1_score(y_true=y_train[valid], y_pred=predictions, zero_division=0)

    # Store the result
    f1s.append(f1)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)

    # Print the scores for each fold
    print(f"Precision = {precision}")
    print(f"Recall = {recall}")
    print(f"Accuracy = {accuracy}")
    print(f"F1 score = {f1}\n")

    fold += 1
```

```python
print("Weights ",model.coef_)
print("Num of Lists: ", len(model.coef_))
weights = model.coef_[0].tolist()  # Assuming model.coef_ is a 2D array
rounded_weights = [f"{weight:.2f}" for weight in weights]
print("Weights list rounded:", rounded_weights)
```

```python
y_pred = model.predict(x_test)
classification = classification_report(y_test, y_pred)

print("\nClassification Report:\n", classification)
```
